```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PowerTransformer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn.preprocessing import MinMaxScaler #MinMaxScaler
from sklearn.preprocessing import StandardScaler #Standard
from sklearn.preprocessing import RobustScaler #Robust
from sklearn.preprocessing import Normalizer #Normal
```

```python
def logTrans(df, feature):   
    
    ft_positiva = df[feature][df[feature]>0]

    #plot da variável sem modificações
    plt.hist(ft_positiva, bins=30, color='blue', edgecolor='black')
    
    # Adicionar rótulos e título
    plt.xlabel('Valores da Feature')
    plt.ylabel('Frequência')
    plt.title(f'Histograma da Feature {feature}')
    
    # Mostrar o histograma
    plt.show()
    
    #Aplicando o log na variável e somando 1 para evitar o problema com log 0
    variavel_log = np.log1p(ft_positiva + 0.0001)
    
    plt.hist(variavel_log, bins=30, color='green', edgecolor='black')
    
    # Adicionar rótulos e título
    plt.xlabel('Valores da Feature')
    plt.ylabel('Frequência')
    plt.title(f'Histograma da Feature {feature} com log aplicado')
    
    # Mostrar o histograma
    plt.show()
    
    variancia_sem_log = ft_positiva.var()
    variancia_com_log = variavel_log.var()

    print('---------------------------------------')
    print(f'variância sem log: {variancia_sem_log}')
    print(f'variância com log: {variancia_com_log}')
    print('---------------------------------------')
```

```python
def sqrTrans(df, feature):   

    #plot da variável sem modificações
    plt.hist(df[feature], bins=30, color='blue', edgecolor='black')
    
    # Adicionar rótulos e título
    plt.xlabel('Valores da Feature')
    plt.ylabel('Frequência')
    plt.title(f'Histograma da Feature {feature}')
    
    # Mostrar o histograma
    plt.show()
    
    variavel_sqr= np.square(df[feature])
    
    plt.hist(variavel_sqr, bins=30, color='green', edgecolor='black')
    
    # Adicionar rótulos e título
    plt.xlabel('Valores da Feature')
    plt.ylabel('Frequência')
    plt.title(f'Histograma da Feature {feature} com sqr aplicado')
    
    # Mostrar o histograma
    plt.show()
    
    variancia_sem_sqr = df[feature].var()
    variancia_com_sqr = variavel_sqr.var()

    print('---------------------------------------')
    print(f'variância sem sqr: {variancia_sem_sqr}')
    print(f'variância com sqr: {variancia_com_sqr}')
    print('---------------------------------------')
```

```python
def boxcox(df, feature):   
    
    boxcoxTr = PowerTransformer(method = "box-cox", standardize=True)

    #plot da variável sem modificações
    plt.hist(df[feature], bins=30, color='blue', edgecolor='black')
    
    # Adicionar rótulos e título
    plt.xlabel('Valores da Feature')
    plt.ylabel('Frequência')
    plt.title(f'Histograma da Feature {feature}')
    
    # Mostrar o histograma
    plt.show()
    
    
    df_boxcox = pd.DataFrame(boxcoxTr.fit_transform(df[feature].values.reshape(-1,1)))
    
    plt.hist(df_boxcox, bins=30, color='green', edgecolor='black')
    
    # Adicionar rótulos e título
    plt.xlabel('Valores da Feature')
    plt.ylabel('Frequência')
    plt.title(f'Histograma da Feature {feature} com box-cox aplicado')
    
    # Mostrar o histograma
    plt.show()
    
    variancia_sem_boxcox = df[feature].var()
    variancia_com_boxcox = df_boxcox.var()

    print('---------------------------------------')
    print(f'variância sem box-cox: {variancia_sem_boxcox}')
    print(f'variância com box-cox: {variancia_com_boxcox}')
    print('---------------------------------------')
```

```python
def padronizacao (df, feature):
    
    #plot da variável sem modificações
    plt.hist(df[feature], bins=30, color='blue', edgecolor='black')
    
    # Adicionar rótulos e título
    plt.xlabel('Valores da Feature')
    plt.ylabel('Frequência')
    plt.title(f'Histograma da Feature {feature}')
    
    # Mostrar o histograma
    plt.show()
    
    #Armazenando em um objeto
    scaler = StandardScaler()

    #Passando a feature para um formato vetorial -1,1
    ft_array = df[feature].values.reshape(-1, 1)
    
    #Aplicando na features
    scaled_feature = scaler.fit_transform(ft_array)
    
    #plot da variável
    plt.hist(scaled_feature, bins=30, color='green', edgecolor='black')

    # Adicionar rótulos e título
    plt.xlabel('Valores da Feature')
    plt.ylabel('Frequência')
    plt.title(f'Histograma da Feature {feature} padronizada')
    
    # Mostrar o histograma
    plt.show()
    
    variancia_sem_padronizacao = df[feature].var()
    variancia_com_padronizacao = scaled_feature.var()

    print('---------------------------------------')
    print(f'variância sem padronização: {variancia_sem_padronizacao}')
    print(f'variância com padronização: {variancia_com_padronizacao}')
    print('---------------------------------------')
```
